# Simple Metrics - Running it back

The goal of this notebook is to take a step back from the more experimental metrics and get a baseline understanding of the two main datasets we're looking at. This is to be done with simple paper count and citation count metrics, as well as by putting the fields of study to the side to avoid complicating things further.

```python
import pandas as pd
import numpy as np
```

```python
bdf = pd.read_csv('/bgfs/mrfrank/nolan/threelistfilterAllPapers.csv')
bdf
```

```python
papers = pd.read_csv('/bgfs/mrfrank/datawarehouse/MAG/NEW_MAG/Papers.txt', sep = '\t', names=[
                                      "Paper ID",
                                      "Rank",
                                      "Doi string",
                                      "Doc type",
                                      "Paper title",
                                      "Original title",
                                      "Book title string",
                                      "Year",
                                      "Date",
                                      #"Online date",
                                      "Publisher",
                                      "Journal ID",
                                      "Conference Series ID",
                                      "Conference instance ID",
                                      "Volume",
                                      "Issue",
                                      "First page",
                                      "Last page",
                                      "Reference count",
                                      "Citation count",
                                      "Estimated citation",
                                      "Original venue",
                                      #"Family ID",
                                      #"Family rank",
                                      "Created Date"])
papers = papers[['Paper ID', 'Paper title', 'Year', 'Date', 'Reference count', 'Citation count']]
```

```python
ap = pd.read_csv('/bgfs/mrfrank/datawarehouse/MAG/NEW_MAG/PaperAuthorAffiliations.txt',
                 sep = '\t', 
                 names = ['Paper ID', 'Author ID', 'Affiliation ID', 'Author sequence number'])
```

```python
affils = pd.read_csv("/bgfs/mrfrank/datawarehouse/MAG/NEW_MAG/Affiliations.txt",sep="\t",header=None,
                                           names=[
                                               "Affiliation ID",
                                               "Rank",
                                               "Affiliation name",
                                               "Display name",
                                               "Paper count",
                                               "Citation count"
                                           ],
                                           index_col=0,
                                           engine='c',
        )
affils = affils[['Affiliation name']]
affils
```

```python
bdfpap = papers[papers['Paper ID'].isin(bdf['Paper ID'])]
```

```python
papaffil = pd.merge(bdfpap, ap, on = 'Paper ID', how = 'inner')
```

```python
paa = pd.merge(papaffil, affils, on = 'Affiliation ID', how = 'left')
paa
```

```python
bfdf = paa[['Paper ID', 'Paper title', 'Year', 'Reference count', 'Citation count', 'Affiliation ID', 'Affiliation name']]
```

```python
bfdf.to_csv('/bgfs/mrfrank/nolan/MAG/Data/BroadDF_Affils.csv')
```
