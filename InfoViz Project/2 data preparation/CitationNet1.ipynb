```python
import pandas as pd
import numpy as np
```

```python
bdf = pd.read_csv('/bgfs/mrfrank/nolan/MAG/Data/BroadDF_Affils.csv', index_col = 0)
```

```python
papers = pd.read_csv('/bgfs/mrfrank/datawarehouse/MAG/NEW_MAG/Papers.txt', sep = '\t', names=[
                                      "Paper ID",
                                      "Rank",
                                      "Doi string",
                                      "Doc type",
                                      "Paper title",
                                      "Original title",
                                      "Book title string",
                                      "Year",
                                      "Date",
                                      #"Online date",
                                      "Publisher",
                                      "Journal ID",
                                      "Conference Series ID",
                                      "Conference instance ID",
                                      "Volume",
                                      "Issue",
                                      "First page",
                                      "Last page",
                                      "Reference count",
                                      "Citation count",
                                      "Estimated citation",
                                      "Original venue",
                                      #"Family ID",
                                      #"Family rank",
                                      "Created Date"])
papers = papers[['Paper ID', 'Paper title', 'Year', 'Date', ]]
```

```python
pr = pd.read_csv('/bgfs/mrfrank/datawarehouse/MAG/NEW_MAG/PaperReferences.txt', names = ['Source ID', 'Target ID'],
                sep = '\t')
pr
```

```python
bdf = bdf[['Paper ID', 'Paper title', 'Year', 'Affiliation ID', 
          'Affiliation name']]
```

```python
bdf = pd.merge(bdf, pr, left_on = 'Paper ID', right_on = 'Target ID', how = 'left')
```

```python
bdf.rename(columns = {'Paper title': 'Target title', 'Year': 'Target year', 'Author ID': 'Target Author ID', 'Author name': 'Target author name',
                    'Affiliation ID': 'Target affiliation ID', 'Affiliation name': 'Target affiliation name', 'Country': 'Target country',
                    'Region': 'Target region'}, inplace = True)
```

```python
bdf = pd.merge(bdf, papers, left_on = 'Source ID', right_on = 'Paper ID', how = 'left')
```

```python
bdf.rename(columns = {'Paper title': 'Source title', 'Year': 'Source year', 'Author ID': 'Source Author ID', 'Author name': 'Source author name',
                    'Affiliation ID': 'Source affiliation ID', 'Affiliation name': 'Source affiliation name', 'Country': 'Source country',
                    'Region': 'Source region'}, inplace = True)
```

```python
bdf.to_csv('/bgfs/mrfrank/nolan/MAG/Data/BroadDF_IncomingCitations.csv')
```
