# Network edges

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
matches = pd.read_csv('/bgfs/mrfrank/nolan/MAG/Data/FuzzyAllAffils.csv', index_col = 0)
matches
```

```python
crdf = pd.read_csv('/bgfs/mrfrank/nolan/MAG/Data/CountryRegionPairs.csv')
crdf
```

```python
bdf = pd.read_csv('/bgfs/mrfrank/nolan/MAG/Data/BroadDF_IncomingCitations.csv', index_col = 0, engine = 'python')
```

```python
bmdf = pd.merge(bdf, matches, left_on = ['Target affiliation name'], right_on = ['Affiliation name'], how = 'left')\
         .drop(['Freq', 'Affiliation name'], axis = 1)\
         .rename(columns = {'Match':'Target match', 'Country':'Target matched country'})
bmdf2 = pd.merge(bmdf, matches, left_on = ['Source affiliation name'], right_on = ['Affiliation name'], how = 'left')\
          .drop(['Freq','Affiliation name'], axis = 1)\
          .rename(columns = {'Match':'Source match', 'Country':'Source matched country'})
```

```python
bmdfs = pd.merge(bmdf2, crdf,left_on = 'Source matched country', right_on = 'Country', how = 'left')\
          .drop('Country', axis = 1).rename(columns = {'Region':'Source matched region'})
bmdfs = pd.merge(bmdfs, crdf,left_on = 'Target matched country', right_on = 'Country', how = 'left')\
          .drop('Country', axis = 1).rename(columns = {'Region':'Target matched region'})
bmdfs
```

```python
bdf2 = bmdfs[['Source ID', 'Source title', 'Source year', 'Source match', 'Source matched country', 'Source matched region', 
             'Target ID', 'Target title', 'Target year', 'Target match', 'Target matched country', 'Target matched region']]\
        .drop_duplicates()
```

```python
bdf2.to_csv('/bgfs/mrfrank/nolan/MAG/Data/Broad_MatchAffilNetwork.csv')
```
